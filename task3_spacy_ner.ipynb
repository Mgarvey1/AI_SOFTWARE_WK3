```python
import pandas as pd

file_path = '/content/drive/MyDrive/AI_Tools_Assignment/practical/train.ft.txt'  # match exact name

with open(file_path, 'r', encoding='utf-8') as f:
    lines = f.readlines()

df = pd.DataFrame({'reviewText': lines})
df.head()
```

```python
!pip install spacy
!python -m spacy download en_core_web_sm

import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
# Extract entities from first 20 reviews
for i, text in enumerate(df['reviewText'][:20]):
    doc = nlp(text)
    ents = [(ent.text, ent.label_) for ent in doc.ents]
    if ents:
        print(f"Review {i}: Entities:", ents)
```

```python
# Simple positive/negative word lists
pos_words = set(["good","great","excellent","amazing","love","awesome","perfect","best"])
neg_words = set(["bad","terrible","awful","hate","worst","poor","disappointing"])

def rule_sentiment(text):
    tokens = [t.lower().strip(".,!?:;") for t in text.split()]
    score = sum(1 for t in tokens if t in pos_words) - sum(1 for t in tokens if t in neg_words)
    if score > 0:
        return "positive"
    elif score < 0:
        return "negative"
    else:
        return "neutral"

# Apply to first 10 reviews
for text in df['reviewText'][:10]:
    doc = nlp(text)
    ents = [(ent.text, ent.label_) for ent in doc.ents]
    sentiment = rule_sentiment(text)
    print("Sentiment:", sentiment, "| Entities:", ents)
```
